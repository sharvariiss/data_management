```python
from pathlib import Path

# Define path to the XML metadata file
xml_path = Path("data/metadata/climate_dataset_dublin_core.xml")

# Show absolute path for verification/debugging
print("Metadata path:", xml_path.resolve())

# Check if file exists before further processing
print("File exists:", xml_path.exists())
```

```python

raw_xml = xml_path.read_text(encoding="utf-8")
print(raw_xml[:1500])
```

```python
import xml.etree.ElementTree as ET

import pandas as pd
```

```python
# parse XML to tree 
tree = ET.parse(xml_path)

# access the root element of the XML tree
root = tree.getroot()

# XML Structure 
print(ET.tostring(root, encoding="unicode")[:1500])

# main tag of XML 
print("Root tag:", root.tag)

# child elements 
print("Child tags:", [child.tag for child in root])

# number of child elements 
print("Number of child elements:", len(root))
```
